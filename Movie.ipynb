##3. Handle Missing Values
Checking for missing values is an important step in data cleaning because missing values can have a significant impact on the quality and reliability of the data analysis. Missing values can lead to incomplete or inaccurate data, which can skew the analysis and produce misleading results. By identifying and handling missing values appropriately, me ensure the integrity and reliability of the data.

```python
#Check the amount of missing values
def nulls(df):
    null_values = pd.DataFrame(df.isnull().sum())
    null_values[1] = null_values[0] / len(df)
    null_values.columns = ['count','%pct']
    filtered_null = null_values[null_values['%pct'] > 0].sort_values(by='%pct', ascending=False)
    return filtered_null
nulls(df)
```

- Strategy One: Drop Column
The percentage of null value is more than 70%.

- Strategy Two: Drop Rows:
The percentage of null value is less than 5%.

- Strategy Three: Impute with Mode:
The percentage of null value is between 5% and 10%.
The datatype is categorical or lable

- Strategy Four: Impute with Mean or Median:
The percentage of null value is between 5% and 10%.
The datatype is numerical

- Strategy Five: Impute with KNN Method:
The percentage of null value is between 10% and 70%.
The datatype is categorical lable
They are not MNAR

- Strategy Six: Impute with KNN Method:
The percentage of null value is between 10% and 70%.
The datatype is Number
They are not MNAR

For NaN values of 17% such as gross, use the sixth strategy, for NaN values between 5% - 10%, such as aspect_ratio and content_rating, use strategies three and four. while for the rest who have NaN values below 5% use strategy two.
To solve the Missing Values case in the form of blank values in the gross column, for example the solution to be chosen is: each blank value will be replaced with 0.

```python
#Remove missing values
df['gross'].fillna(0, inplace=True)
value_mean = df['aspect_ratio'].mean()
df = df.fillna({'aspect_ratio': value_mean})
df['content_rating'] = df['content_rating'].fillna(df['content_rating'].mode()[0])
df = df.dropna(subset = ['plot_keywords', 'title_year', 'director_facebook_likes', 'director_name',
                         'num_critic_for_reviews', 'actor_3_name', 'actor_3_facebook_likes',
                         'num_user_for_reviews', 'color', 'duration', 'facenumber_in_poster',
                         'actor_2_name', 'actor_2_facebook_likes', 'language', 'actor_1_name',
                         'actor_1_facebook_likes', 'country'])
```

##1. Check Possible Range
Checking the possible range is an important step in data cleaning because it helps identify and handle data values that fall outside the expected or valid range for a particular variable. By examining the range of values, we can identify outliers, data entry errors, or any other data points that are unlikely or impossible given the context of the data.

```python
#Shows the possible range
num_var = [col for col in df.columns if df[col].dtypes != 'O']
num_var
```

```python
#Check possible range
for i in range(len(num_var)):
    per = round((len(df[df[num_var[i]] < 0]) / len(df)) * 100, 5)
    print('{} | min {} | max {} | percentage %{}'.format(num_var[i], df[num_var[i]].min(), df[num_var[i]].max(), per))
```

##2. Sanity Check
Sanity checks can help ensure the consistency and integrity of the data. If two numerical variables are expected to have a certain relationship or dependency, a sanity check can help to understand. Among all numerical variables, first, we must realize there is a logical relationship between which variables. We should analye the relationship based on the data documentation.

- sanity 1 : actor_1_facebook_likes <= cast_total_facebook_likes
- sanity 2 : actor_2_facebook_likes <= cast_total_facebook_likes
- sanity 3 : actor_3_facebook_likes <= cast_total_facebook_likes

```python
#Sanity checks
df['Sanity1'] = ['OK' if x <= y else 'NOT' for x, y in zip(df['actor_1_facebook_likes'], df['cast_total_facebook_likes'])]
df['Sanity2'] = ['OK' if x <= y else 'NOT' for x, y in zip(df['actor_2_facebook_likes'], df['cast_total_facebook_likes'])]
df['Sanity3'] = ['OK' if x <= y else 'NOT' for x, y in zip(df['actor_3_facebook_likes'], df['cast_total_facebook_likes'])]
```

#Adding Columns

Adding a category duration column, this is only optional.

```python
#Adding a category duration column
def category_duration(duration):
    if (duration<90):
      return 'Category_1'
    elif ((duration>=90) and (duration<=120)):
      return 'Category_2'
    else:
      return 'Category_3'
```
