```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Read Data

```python
#Import dataset
sheet_url = 'https://docs.google.com/spreadsheets/d/1bK36H8X5jgfDacwKDX9F3tNOaSm6s7aNAw44gIuWNzs/edit#gid=620735373'
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
df = pd.read_csv(sheet_url_trf)
pd.set_option('display.max_columns', None)
df.head()
```

#Missing Value

```python
#see NaN value
df.isna().sum()
```

#Replacing NaN Value

```python
#example of duration column
df['duration'].head()
```

```python
#change the NaN value using the mean value
value_mean = df['duration'].mean()
df = df.fillna({'duration': value_mean})
df['duration'].head()
```

#Adding Columns

```python
#Add a column with conditions
def category_duration(duration):
    if (duration<90):
      return 'Category_1'
    elif ((duration>=90) and (duration<=120)):
      return 'Category_2'
    else:
      return 'Category_3'
```

```python
#create columns
df['movie_duration_category'] = df['duration'].apply(category_duration)
df_view = df[['movie_title', 'duration', 'movie_duration_category']]
df_view
```
