##2. Read Data

```python
import io
df=pd.read_csv(io.StringIO(filenya['TheLook Ecommerce.csv'].decode('utf-8')))
print(df)

#melihat isi data
```

## 3. Change The Column Name and Data Type

```python
df['created_at']=pd.to_datetime(df['created_at'])
df['returned_at']=pd.to_datetime(df['returned_at'])
df['shipped_at']=pd.to_datetime(df['shipped_at'])
df['delivered_at']=pd.to_datetime(df['delivered_at'])
df['returned_at_1']=pd.to_datetime(df['returned_at_1'])
df['created_at_1']=pd.to_datetime(df['created_at_1'])
```

```python
df['order_id'] = df.order_id.astype(object)
df['user_id'] = df.user_id.astype(object)
df['id'] = df.id.astype(object)
df['age'] = df.age.astype(object)
```
